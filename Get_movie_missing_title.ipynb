```python
import requests
import pandas as pd
from imdb_password import api_key
```

```python
# Load your dataset
df_missing_year = pd.read_csv('./DataSets/Movie_missing_year.csv')
```

```python
# Set up the API key and base URL
base_url = 'http://www.omdbapi.com/'
```

```python
# Create a function to fetch the year from OMDb API
def fetch_year_from_omdb(title):
    params = {
        't': title,
        'apikey': api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data.get('Response') == 'True' and 'Year' in data:
            return data['Year']
    return None
```

```python
# Apply the function to fill in missing years
df_missing_year['YEAR'] = df_missing_year['Title'].apply(fetch_year_from_omdb)
```
